```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
# Generate x-values from 0 to 10
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, legend_label="sin(x)", line_width=2)
(show(p))
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns


np.random.seed(0)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
sizes = np.random.uniform(5, 20, 100)
colors = np.random.choice(['red', 'green', 'blue', 'yellow'], 100)

# Create a Bokeh figure
p = figure(title="Scatter Plot with Varying Marker Sizes and Colors",
           x_axis_label='x', y_axis_label='y')
p.circle(x, y, size=sizes, color=colors, fill_alpha=0.5)
(show(p))
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')

p.vbar(x=fruits, top=counts, width=0.9, legend_label="Fruit Counts", line_color='white')
(show(p))
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()))

p.image(image=[data_heatmap], x=x.min(), y=y.min(), dw=x.max()-x.min(), dh=y.max()-y.min(),
        palette='Magma256')

show(p)
```
